```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pylab as pl
```

```python
taxi1 = pd.read_csv('/content/taxi_owners.csv')
taxi2 = pd.read_csv('/content/taxi_vehicles.csv')
```

```python
taxi1.head()
```

```python
taxi2.head()
```

```python
merged_df = pd.merge(taxi1, taxi2, on='vid')

print(merged_df.head())
```

```python
merged_df.info()
```

```python
print(merged_df['fuel_type'].value_counts())
```

```python
# Statistik deskriptif untuk kolom numerik
print(merged_df.describe())

# Statistik deskriptif untuk kolom kategorikal
print(merged_df.describe(include=['object']))
```

```python
print(merged_df['make'].value_counts())

# Frekuensi jenis bahan bakar
print(merged_df['fuel_type'].value_counts())
```

```python
# Count plot untuk jenis bahan bakar
sns.countplot(data=merged_df, x='fuel_type')
plt.title('Jumlah Kendaraan Berdasarkan Jenis Bahan Bakar')
plt.show()
```

```python
# Buat histogram
plt.figure(figsize=(12, 6))
sns.histplot(data=merged_df, x="year", bins=20, kde=True, color="skyblue")
plt.title("Distribusi Umur Kendaraan Taksi", fontsize=16, fontweight='bold')
plt.xlabel("Tahun Pembuatan")
plt.ylabel("Densitas")
plt.legend()
plt.show()
```

Banyak kendaraan yang dipakai supir taxi dibuat oada tahun 2014

```python
merged_df.info()
```

```python
merged_df.duplicated().sum()
```

```python
count_by_make = merged_df['make'].value_counts()

# Buat plot
plt.figure(figsize=(12, 6))
sns.countplot(x="make", data=merged_df, order=count_by_make.index, palette="Set2")
plt.title("Distribusi Merek Kendaraan Taksi", fontsize=16, fontweight='bold')
plt.xlabel("Merek Kendaraan")
plt.ylabel("Jumlah Kendaraan")
plt.xticks(rotation=45)

# Tambahkan anotasi jumlah kendaraan di atas setiap bar
for p in plt.gca().patches:
    plt.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=11, color='black')

plt.show()
```

Toyota menjadi merek kendaraan taxi paling populer digunakan oleh supir taxinya

```python
sns.countplot(data=merged_df, x='make', hue='fuel_type')
plt.title('Hubungan Antara Merek dan Jenis Bahan Bakar')
plt.xlabel('Merek')
plt.ylabel('Jumlah Kendaraan')
plt.xticks(rotation=45)
plt.show()
```

Toyota banyak memakai bahan bakar Hybrid untuk merek mobilnya

```python
sns.boxplot(x='fuel_type', y='year', data=merged_df)
plt.title('Distribusi Tahun Kendaraan Berdasarkan Jenis Bahan Bakar')
plt.xlabel('Jenis Bahan Bakar')
plt.ylabel('Tahun Kendaraan')
plt.show()
```

Banyak kendaraan buatan tahun 2017 memakai gasoline
