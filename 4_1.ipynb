```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates

# Membaca dataset
data_2015 = pd.read_csv('/content/2015.csv')

# Memilih kolom-kolom yang relevan untuk plot
columns_to_plot = [
    'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)',
    'Family', 'Health (Life Expectancy)', 'Freedom',
    'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual'
]

# Membuat Parallel Coordinates Plot
plt.figure(figsize=(12, 6))
parallel_coordinates(data_2015[columns_to_plot], class_column='Happiness Rank', colormap='rainbow')
plt.title('Parallel Coordinates Plot for World Happiness Report 2015')
plt.xlabel('Attributes')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates

data_2015 = pd.read_csv('/content/2015.csv')

columns_to_plot = [
    'Region', 'Happiness Score', 'Economy (GDP per Capita)',
    'Family', 'Health (Life Expectancy)', 'Freedom',
    'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual'
]

plt.figure(figsize=(14, 8))
parallel_coordinates(data_2015[columns_to_plot], class_column='Region', colormap='Set2')
plt.title('Parallel Coordinates Plot for World Happiness Report 2015 by Region')
plt.xlabel('Attributes')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
